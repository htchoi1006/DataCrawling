```python
from selenium import webdriver
from bs4 import BeautifulSoup
import urllib.request as ur
import random
import time
import json
```

```python
crawling_id = 'shirts_bad_list_27.json'
```

```python
with open(crawling_id) as d:
    dataids = json.load(d)
print(len(dataids))
```

```python
def getFeature(id_list, bad_list, good_list):
    options = webdriver.ChromeOptions()
    driver = webdriver.Chrome('./chromedriver')
    driver.implicitly_wait(1)
    driver.get('https://www.amazon.com/ref=nav_logo?language=en_US')
    
    for i in range(len(id_list)):
        id = id_list[i]
        gender = 'male'
        info = {}
        url = 'https://www.amazon.com/dp/' + id
        print(url)
        try:
            driver.get(url)
            html = driver.page_source
            soup = BeautifulSoup(html, 'html.parser')

            title = soup.select('#productTitle')[0].text.strip()

            category_select = soup.select('#wayfinding-breadcrumbs_feature_div > ul > li > span')
            category = []
            for single_category in category_select:
                category.append(single_category.text.strip())

            feature_select = soup.select('#feature-bullets > ul > li > span')
            feature = []
            for single_feature in feature_select:
                feature.append(single_feature.text.strip())

            imgurl_select = soup.select('#landingImage')[0]['data-a-dynamic-image']
            imgurl = (imgurl_select.split('\"')[1])

            info = {
                'id' : id,
                'title' : title,
                'category' : category,
                'features' : feature,
                'gender' : gender,
                'img_links' : imgurl
            }
            if i % 50 == 0:
                print(str(i) + ' th crawling data')
                print(info)
            
            good_list.append(info)
            if i % 1000 == 0:
                json.dump(good_list, open('./amazon_shirts_28' + str(i) + '.json', 'w+'))
        except:
            bad_list.append(id)
    
```

```python
bad_list = []
good_list = []
getFeature(test, bad_list, good_list)
json.dump(good_list, open('./amazon_shirts_all_28.json', 'w+'))
```
